```python
import os

import matplotlib.pyplot as plt
import torch
from monai import transforms
from monai.config import print_config
from monai.data import DataLoader, CacheDataset
from monai.utils import first, set_determinism
from torch.cuda.amp import GradScaler, autocast
from tqdm import tqdm

from generative.inferers import ControlNetLatentDiffusionInferer
from generative.networks.schedulers import DDPMScheduler

print_config()
```

```python
# for reproducibility purposes set a seed
set_determinism(42)
```

```python
import pandas as pd

df_split = pd.read_csv('TrainValid_split.csv', header=0)
df_split.columns = ["image", "class", "group"]
df_split["label"] = pd.Series(["./2_Train,Valid_Mask/"]*len(df_split)).str.cat(df_split["image"].str.cat(["_label.nii.gz"]*len(df_split)))
df_split["image"] = pd.Series(["./1_Train,Valid_Image/"]*len(df_split)).str.cat(df_split["image"].str.cat([".nii.gz"]*len(df_split)))
train_datalist = df_split[df_split["group"] == "Train"][["image", "label"]].to_dict('records')
train_datalist[0]
```

```python
train_transforms = transforms.Compose(
    [
        transforms.LoadImaged(keys=["image", "label"]),
        transforms.EnsureChannelFirstd(keys=["image", "label"]),
        transforms.EnsureTyped(keys=["image", "label"]),
        transforms.Orientationd(keys=["image", "label"], axcodes="RAS"),
        transforms.Spacingd(keys=["image", "label"], pixdim=(2.0, 2.0, 3.2), mode=("bilinear", "nearest")), # 2.0, 2.0, 3.2
        transforms.CenterSpatialCropd(keys=["image", "label"], roi_size=(128, 128, 128)), # As big as your GPU can handle
        transforms.SpatialPadd(keys=["image", "label"], spatial_size=(128, 128, 128)), # In our tests, we used 128x128x128
        transforms.ScaleIntensityRangePercentilesd(keys="image", lower=0, upper=99.5, b_min=0, b_max=1),
    ]
)

batch_size = 1 # Set to one for inference

train_ds = CacheDataset(
    data=train_datalist,
    transform=train_transforms,
    cache_rate=1.0,
    num_workers=2,
)

train_loader = DataLoader(train_ds, batch_size=batch_size, shuffle=True, num_workers=2, persistent_workers=True)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using {device}")
```

```python
autoencoder = torch.load("autoencoder_control.pth")
autoencoder.to(device)
autoencoder.eval()
unet = torch.load("unet_control.pth")
unet.to(device)
unet.eval()
controlnet = torch.load("controlnet_control.pth")
controlnet.to(device)
controlnet.eval()

scale_factor = 0.9982640147209167 # From my training

scheduler = DDPMScheduler(num_train_timesteps=1000, schedule="scaled_linear_beta", beta_start=0.0015, beta_end=0.0195)
```

```python
controlnet_inferer = ControlNetLatentDiffusionInferer(scheduler)
```

```python
for batch in train_loader:
    with torch.no_grad():
        with autocast(enabled=True):
            z = autoencoder.encode_stage_2_inputs(batch["image"].to(device))
            noise = torch.randn_like(z).to(device) # Noise needs to be the same shape as the autoencoder output
            sample = controlnet_inferer.sample(
                input_noise = noise,
                diffusion_model = unet,
                autoencoder_model = autoencoder,
                controlnet = controlnet,
                cn_cond = batch["label"].to("cuda"), # Segmentation
                scheduler = scheduler,
            )

    img = sample[0, 0].cpu()
    lbl = batch["label"][0, 0].cpu()

    fig, axs = plt.subplots(nrows=2, ncols=3)
    for axss in axs:
        for ax in axss:
            ax.axis("off")
    ax = axs[0][0]
    ax.imshow(img[..., img.shape[2] // 2], cmap="gray")
    ax = axs[0][1]
    ax.imshow(img[:, img.shape[1] // 2, ...], cmap="gray")
    ax = axs[0][2]
    ax.imshow(img[img.shape[0] // 2, ...], cmap="gray")
    ax = axs[1][0]
    ax.imshow(lbl[..., img.shape[2] // 2], cmap="gray")
    ax = axs[1][1]
    ax.imshow(lbl[:, img.shape[1] // 2, ...], cmap="gray")
    ax = axs[1][2]
    ax.imshow(lbl[img.shape[0] // 2, ...], cmap="gray")
    plt.show()
```
